## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
#merging both data
combine_data = pd.merge(mouse_metadata, study_results, on = "Mouse ID", how = "outer")
combine_data.head()

```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_data = combine_data.groupby("Drug Regimen")

summary_mean = summary_data["Tumor Volume (mm3)"].mean()
summary_median = summary_data["Tumor Volume (mm3)"].median()
summary_variance = summary_data["Tumor Volume (mm3)"].var()
summary_std = summary_data["Tumor Volume (mm3)"].std()
summary_sem = summary_data["Tumor Volume (mm3)"].sem()

summary_table = pd.DataFrame({"Mean": summary_mean, "Median": summary_median, "Variance": summary_variance, "Standard Deviation": summary_std, "SEM": summary_sem})

summary_table
```

## Bar plots

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
#to count number of data points group by drug regimen
summary_count = combine_data.groupby(["Drug Regimen"]).count()["Mouse ID"]

#make bar plot
summary_count.plot(kind="bar", figsize=(10,5))

#set names for title and x/y axis
plt.title("bar plot: data points of drug regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("data points")

#show graph
plt.show()
plt.tight_layout()
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 

#find regimens data
capo = combine_data.loc[combine_data["Drug Regimen"] == "Capomulin",:]
rami = combine_data.loc[combine_data["Drug Regimen"] == "Ramicane",:]
infu = combine_data.loc[combine_data["Drug Regimen"] == "Infubinol",:]
ceft = combine_data.loc[combine_data["Drug Regimen"] == "Ceftamin",:]


#Capomulin


quartiles_capo = capo["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq_capo = quartiles_capo[0.25]
upperq_capo = quartiles_capo[0.75]
iqr_capo = upperq_capo-lowerq_capo

#print capomulin
print(f"The lower quartile of Capomulin is: {lowerq_capo}")
print(f"The upper quartile of Capomulin is: {upperq_capo}")
print(f"The interquartile range of Capomulin is: {iqr_capo}")
print(f"The the median of Capomulin is: {quartiles_capo[0.5]} ")

lower_bound_capo = lowerq_capo - (1.5*iqr_capo)
upper_bound_capo = upperq_capo + (1.5*iqr_capo)

print(f"Values below {lower_bound_capo} could be outliers.")
print(f"Values above {upper_bound_capo} could be outliers.")
print(f"")


#Ramicane
quartiles_rami = rami["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq_rami = quartiles_rami[0.25]
upperq_rami = quartiles_rami[0.75]
iqr_rami = upperq_rami-lowerq_rami


#print Ramicane
print(f"The lower quartile of Ramicane is: {lowerq_rami}")
print(f"The upper quartile of Ramicane is: {upperq_rami}")
print(f"The interquartile range of Ramicane is: {iqr_rami}")
print(f"The the median of Ramicane is: {quartiles_rami[0.5]} ")

lower_bound_rami = lowerq_rami - (1.5*iqr_rami)
upper_bound_rami = upperq_rami + (1.5*iqr_rami)

print(f"Values below {lower_bound_rami} could be outliers.")
print(f"Values above {upper_bound_rami} could be outliers.")
print(f"")



#Infubinol
quartiles_infu = infu["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq_infu = quartiles_infu[0.25]
upperq_infu = quartiles_infu[0.75]
iqr_infu = upperq_infu-lowerq_infu


#print Infubinol
print(f"The lower quartile of Infubinol is: {lowerq_infu}")
print(f"The upper quartile of Infubinol is: {upperq_infu}")
print(f"The interquartile range of Infubinol is: {iqr_infu}")
print(f"The the median of Infubinol is: {quartiles_infu[0.5]} ")

lower_bound_infu = lowerq_infu - (1.5*iqr_infu)
upper_bound_infu = upperq_infu + (1.5*iqr_infu)

print(f"Values below {lower_bound_infu} could be outliers.")
print(f"Values above {upper_bound_infu} could be outliers.")
print(f"")

#Ceftamin
quartiles_ceft = ceft["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq_ceft = quartiles_ceft[0.25]
upperq_ceft = quartiles_ceft[0.75]
iqr_ceft = upperq_ceft-lowerq_ceft


#print ceftamin
print(f"The lower quartile of Ceftamin is: {lowerq_ceft}")
print(f"The upper quartile of Ceftamin is: {upperq_ceft}")
print(f"The interquartile range of Ceftamin is: {iqr_ceft}")
print(f"The the median of Ceftamin is: {quartiles_ceft[0.5]} ")

lower_bound_ceft = lowerq_ceft - (1.5*iqr_ceft)
upper_bound_ceft = upperq_ceft + (1.5*iqr_ceft)

print(f"Values below {lower_bound_ceft} could be outliers.")
print(f"Values above {upper_bound_ceft} could be outliers.")
print(f"")

#set table
table_com = pd.DataFrame({
"lower quartile": [lowerq_capo, lowerq_rami, lowerq_infu, lowerq_ceft],
"upper quartile": [upperq_capo, upperq_rami, upperq_infu, upperq_ceft],
"interquartile range": [iqr_capo, iqr_rami, iqr_infu, iqr_ceft],
"the median": [quartiles_capo[0.5], quartiles_rami[0.5], quartiles_infu[0.5] ,quartiles_ceft[0.5]],
"Values below for outliers": [lower_bound_capo, lower_bound_rami, lower_bound_infu, lower_bound_ceft],
"Values above for outliers": [upper_bound_capo,  upper_bound_rami,  upper_bound_infu,  upper_bound_ceft]
})


table_com.index="Capomulin", "Ramicane", "Infubinol", "Ceftamin"


table_com



```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#time pint data and tumor of capo
timepoint = capo["Timepoint"]
tumor = capo["Tumor Volume (mm3)"]

#set line plot
line_plot = plt.plot(timepoint, tumor)
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume')
plt.title('Tumor Volume vs Time for mouse with capomulin')
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
#mouse avg weight and tumor
mouse_weight = capo.groupby(capo["Mouse ID"])["Weight (g)"].mean()
avg_tumor = capo.groupby(capo["Mouse ID"])["Tumor Volume (mm3)"].mean()

#set scatter plot
plt.scatter(mouse_weight, avg_tumor)
plt.xlabel("Weight")
plt.ylabel("Tumor")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(mouse_weight, avg_tumor)

fit = slope * mouse_weight + intercept

plt.scatter(mouse_weight, avg_tumor)
plt.plot(mouse_weight,fit,"--")
plt.xticks(mouse_weight)
plt.xlabel("Weight")
plt.ylabel("Tumor")
plt.show()
```
